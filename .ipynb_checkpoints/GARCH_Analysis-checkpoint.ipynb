# Volatility Forecasting with a GARCH(1,1) Model

Objective: This notebook implements a GARCH(1,1) model to forecast the daily volatility of the S&P 500 (SPY). Building on our previous analysis which rejected the assumption of normally distributed, constant-variance returns, we now address the phenomenon of volatility clustering. The objective is to quantify, model, and visualise this time-varying volatility, providing a more realistic foundation for risk management and asset pricing.

# 1. Theoretical Motivation & Data Analysis

Financial returns are not independent and identically distributed (i.i.d). While the returns themselves exhibits very little serial correlation, their volatility is highly persistent. Large price changes tend to be followed by large changes, and small changes by small changes.

# 1.1 Confirming Volatility Clustering

To confirm this, we examine the autocorrelation of the squared returns, which serve as a proxy for the daily variance. If volatility is persistent, squared returns should be significantly correlated with their own past values.

```python
# Core data science and statistics libraries
import pandas as pd
import numpy as np
from arch import arch_model

# Visualization libraries

import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Set a professional plot style
plt.style.use('seaborn-v0_8-whitegrid')

# Load SPY log returns from the previous project
# For reproducibility, we'll fetch it again.
import yfinance as yf
# Fetch data
spy_df = yf.download('SPY', start='2005-01-01', end='2025-01-01', progress=False)
spy_df = spy_df[['Close']]

# Compute log returns
spy_df['log_return'] = 100 * np.log(spy_df['Close'] / spy_df['Close'].shift(1))
spy_df.dropna(inplace=True)

# Extract returns series
returns = spy_df['log_return']
```

```python
# Plot ACF and PACF of squared returns
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

plot_acf(returns**2, lags=40, ax=ax1, title='Autocorrelation Function (ACF) of Squared Returns')
ax1.set_xlabel("Lag")
ax1.set_ylabel("Autocorrelation")


plot_pacf(returns**2, lags=40, ax=ax2, title='Partial Autocorrelation Function (PACF) of Squared Returns')
ax2.set_xlabel("Lag")
ax2.set_ylabel("Partial Autocorrelation")

plt.tight_layout()
plt.show()
```

Observation: The ACF plot shows a slow decay, and the PACF plot shows a significant spike at the first lag. This is the classic signature of an ARMA-like process in the variance. The statistically significant autocorrelations confirms the presence of volatility clutering and justify the use of a GARCH model.

# 2. The GARCH(1,1) Model

The GARCH (Generalised Autoregressive Conditional Heteroskedasticity) model formalises this observation. A GARCH(1,1) model specifies the conditional variance sigma_t^2 as a function of three components:

\(\sigma _{t}^{2}=\omega +\alpha r_{t-1}^{2}+\beta \sigma _{t-1}^{2}\)

Where:
- sigma_t^2 is the conditional variance in this case.
- omega is the long-run average variance (a constant).
- r_t - 1^2 is yesterday's swuared return (the ARCH term), representing the "news" or shock from the previous period. The coefficient (alpha) governs our reaction to this news.
- sigma_t - 1^2 is yesterday's conditional variance (the GARCH term), representing the persistent, "sticky" component of volatility. The coefficient (beta) governs this persistence.

For a stable model, we require alpha + beta \<1

# 2.1 Model Implementation

We fit a GARCH(1,1) model with a constant mean and assume normally distributed innovations for this first implementation.

```python
# Specify the GARCH(1,1) model
# We use a constant mean, p=1 (ARCH), q=1 (GARCH)
garch_model = arch_model(returns, vol='Garch', p=1, q=1,
                         mean='Constant', dist='Normal')

# Fit the model
results = garch_model.fit(disp='off')

# Print the detailed summary
print(results.summary())
```

# 3. Interpretation of Results

# 3.1 Key Financial Insights

- Volatility Persistence: The sum: alpha + beta = 0.1361 + 0.8493 = 0.9854. This value is very close to 1, indicating that shocks to volatility are highly persistent. It takes a long time for volatility to revert to its long-term mean after a market shock.
- Volatility Half-Life: We can quantify this persistence by calculating the volatility half-life, which is the time it takes for a shock's impact to decay by 50%. $$ \tau_{1/2} = \frac{\ln(0.5)}{\ln(\alpha + \beta)} $$

```python
alpha = results.params['alpha[1]']
beta = results.params['beta[1]']
half_life = np.log(0.5) / np.log(alpha + beta)
print(f"Volatility Persistence (alpha + beta): {alpha + beta:.4f}")
print(f"Volatility Half-Life: {half_life:.2f} days")
```

This means a major shock to the market (like a geopolitical event or earnings surprise) will still have half of its impact on volatility approximately 58 days later. This is a critical insight for risk management.

# 4. Visualisation of Conditional Volatility

Plotting the model's output against the daily returns shows how effectively the GARCH model captures the periods of high and low volatility.

```python
# Get the conditional volatility from the model
conditional_vol = results.conditional_volatility

# Plotting the results
plt.figure(figsize=(14, 7))
# Plot returns
plt.plot(returns.index, returns, color='grey', alpha=0.6, label='Daily Log Returns (%)')
# Plot conditional volatility
plt.plot(conditional_vol.index, conditional_vol, 'r-', label='GARCH(1,1) Conditional Volatility')

plt.title('SPY Daily Returns and GARCH Conditional Volatility', fontsize=14)
plt.ylabel('Percentage (%)')
plt.xlabel('Date')
plt.legend()
plt.show()
```

The model successfully adapts, with conditional volatility rising sharply during the 2008 Financial Crisis and the 2020 COVID-19 crash, and subsiding during calmer periods.

# 5. Model Diagnostics

A good model should produce residuals that are close to white noise. We check the standardised residuals (r_t/sigma_t) to see if our model has successfully captured the dynamics of volatility. If the model is well-specified, the standardised residuals should not exhibit any remaining autocorrelation or volatility clustering.

```python
std_resid = results.std_resid
# Plot ACF of standardized residuals
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
plot_acf(std_resid**2, lags=40, ax=ax, title='ACF of Standardized Squared Residuals')
plt.show()
```

Observation: The ACF plot of the squared standardised residuals shows that most of the autocorrelation has been successfully removed. The significantly spikes we saw in the original squared returns are now largely gone, indicating that the GARCH(1,1) model is a good fit for the data

# 6. Conclusion and Limitations

This analysis successuly demonstartes that a GARCH(1,1) model can effectively capture the volatility clustering inherent in SPY daily returns. We quantified the high persistent of market volatility and visualised how the model adapts to different market regimes.

Limitations & Next Steps:
1. Symmetry (Leverage Effect): This model assumes positive and negative shocks have the same effect on volatility. In reality, negative shocks (fear) tend to increase volatility more than positive shocks (greed). This could be addressed with an asymmetric GARCH model like GJR-GARCH.
2. Distribution Assumption: We assumed normal innovations, but we know financial returns have fat tails. A better fit could be achieved by using a Student's t-distribution for the innovations (dist='t' in arch_model).

This foundational model provides a robust starting point for more advanced volatility forecasting, which is a cornerstone of quantitative risk management, derivatives pricing, and strategy development.
